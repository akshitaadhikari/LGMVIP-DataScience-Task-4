```python
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Upload an image
uploaded = files.upload()

# Read the uploaded image
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_gray_image = cv2.bitwise_not(gray_image)

# Apply Gaussian blur to the inverted grayscale image
blurred_image = cv2.GaussianBlur(inverted_gray_image, (111,111),0)

# Invert the blurred image
inverted_blurred_image = cv2.bitwise_not(blurred_image)

# Create the pencil sketch by blending the inverted blurred image with the original image
pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
```

```python

# Display the original and pencil sketch images
cv2_imshow(image)
cv2_imshow(pencil_sketch)

# Save the pencil sketch image
sketch_output_path = 'pencil_sketch.jpg'
cv2.imwrite(sketch_output_path, pencil_sketch)

# Download the pencil sketch image
files.download(sketch_output_path)
```
